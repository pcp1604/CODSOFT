# Splitting DATA 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

# Models Training

# <font color="pink">1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score,classification_report
```

```python
log_reg=LogisticRegression()
```

```python
log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
log_reg = accuracy_score(y_test,y_pred)
```

```python
log_reg
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
import scikitplot as skplt
```

```python
skplt.metrics.plot_confusion_matrix(y_test,y_pred,normalize=True)
```

```python
model_result={}
```

```python
model_result["Logistic Regression "]=accuracy_score(y_test,y_pred)
```

```python
model_result
```

# <font color = "purpel">2) DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dct_cls=DecisionTreeClassifier()
dct_cls.fit(x_train,y_train)
```

```python
y_pred=dct_cls.predict(x_test)
```

```python
dct_acc=accuracy_score(y_test,y_pred)
dct_acc
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
skplt.metrics.plot_confusion_matrix(y_test,y_pred,normalize=True)
```

```python
model_result["Decision Tree"]=accuracy_score(y_test,y_pred)
```

```python
model_result
```

# <font color="Red">3) BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag=BaggingClassifier(DecisionTreeClassifier())   
bag.fit(x_train,y_train)                          
```

```python
y_pred_bag=bag.predict(x_test)
```

```python
bag_acc=accuracy_score(y_test,y_pred_bag)
bag_acc
```

```python
print(confusion_matrix(y_test,y_pred_bag))
```

```python
print(classification_report(y_test,y_pred_bag))
```

```python
skplt.metrics.plot_confusion_matrix(y_test,y_pred_bag,normalize=True)
```

```python
model_result["Bagging Classifier"]=accuracy_score(y_test,y_pred_bag)
```

```python
model_result
```

```python
models=pd.DataFrame({
    "Model" : ["Logistic Regression ","DecisionTreeClassifier","BaggingClassifier"],
    "Scores" : [log_reg,dct_acc,bag_acc]
})
```

```python
models
```

```python
colors=["hotpink","tomato","skyblue"]
plt.figure(figsize=(16,8))
sns.barplot(x=models["Model"],y=models["Scores"],palette=colors)
plt.show()
```
